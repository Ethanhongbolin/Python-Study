# For Loop with lists


```python
fruits = ["banana", "mango", "strawberry"]
for fruit in fruits: 
    print(fruit) #prints each elements and add a new line for every element

```

# For Loop with Range

```python
for num in [2, 5, 6, 4]:
    print(num)
```

```python
for number in range(1, 10): #10 is never included, only integers
    print(number)
```

```python
for number in range(1, 10, 2): #10 is never included, only integers, the third number tells how much it goes up by, starts at 1.
    print(number)
```

# Sum Even Number (1-100)

```python
even_number = 0
for number in range(2, 101, 2):
    even_number = even_number + number
print(even_number)
```

```python
target = int(input())
even_sum = 0
for number in range(1, target+1):
    if number % 2 ==0:
        even_sum += number # this is equal to even_sum = even_sum + number

print(even_sum)
```

# FizzBuzz Challenge (1-50)

- number divisible by three with the word “fizz”, 
- and any number divisible by five with the word “buzz”
- and if number is divisible by both three and five, printout "fizzbuzz"

```python
for numbers in range(1, 51):
    if numbers % 3 == 0 and numbers % 5 == 0:
        print(f'number {numbers} is FizzBuzz')
    elif numbers % 3 == 0:
        print(f'number {numbers} is fizz')
    elif numbers % 5 == 0:
        print(f'number {numbers} is Buzz')
    else:
        print(numbers)
```
