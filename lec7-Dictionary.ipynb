# 09. Iterating Over Dictionaries
## Looping Over Keys

```python
programming_dictionary = {
    "Bug": "An error in a program",
    "Function": "A reusable piece of code",
    "Loop": "Doing something repeatedly"
                         }
for key in programming_dictionary:
    print(key)
# Output:
# Bug
# Function
# Loop
```

## Looping Over Values

```python
for value in programming_dictionary.values():
    print(f'these are the values: "{value}"')
# Output:
# An error in a program
# A reusable piece of code
# Doing something repeatedly
```

## Looping Over Key-Value Pairs

```python
for key, value in programming_dictionary.items():
    print(f"{key}: {value}")
# Output:
# Bug: An error in a program
# Function: A reusable piece of code
# Loop: Doing something repeatedly
```

## Dictionary Unpacking with **

```python
def greet(name, age, city):
    print(f"Hi, I'm {name}, {age} years old, from {city}.")
info = {"name": "Alice","age": 30,"city": "New York"}
greet(**info)


# Output: Hi, I'm Alice, 30 years old, from New York.

```

## Enumerate with Dictionaries

```python
for index, (key, value) in enumerate(programming_dictionary.items()):
    print(f"{index}: {key} = {value}")
# Output:
# 0: Bug = An error in a program
# 1: Function = A reusable piece of code
# 2: Loop = Doing something repeatedly
```

# 10 - Dictionary Compreshension

## Basic Syntax

```python
# {key_expression: value_expression for item in iterable}
squares = {x: x ** 2 for x in range(1, 6)}
print(squares)
# Output: {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
```

## From Another Dictionary

```python
prices = {"apple": 1.20,
    "banana": 0.50,
    "cherry": 2.00,
    "date": 3.50
         }
# Double all prices
doubled = {fruit: price * 2 for fruit, price in prices.items()}
print(doubled)
# Output: {'apple': 2.4,'banana': 1.0,'cherry': 4.0,'date': 7.0}
```

## Filtering with a Condition

```python
# Only keep items that cost more than $1
expensive = {fruit: price for fruit, price in prices.items() if price > 1.00}
print(expensive)
# Output: {'apple': 1.2,'cherry': 2.0,'date': 3.5}
```

## Transforming Keys and Values

```python
# Swap keys and values
original = {"a": 1,"b": 2,"c": 3}
swapped = {v: k for k, v in original.items()}
print(swapped)
# Output: {1: 'a', 2: 'b', 3: 'c'}
# Uppercase keys
data = {"name": "Alice","city": "New York"}
upper_keys = {k.upper(): v for k, v in data.items()}
print(upper_keys)
# Output: {'NAME': 'Alice','CITY': 'New York'}

```

## A4 - Vote Tall

```python
votes = [
    "Alice",
    "Bob",
    "Alice",
    "Charlie",
    "Bob",
    "Alice",
    "Alice",
    "Charlie",
    "Bob",
    "Alice",
    "Charlie",
    "Charlie",
]
tally = {}
for name in votes:
    tally[name] = tally.setdefault(name, 0) + 1
    

total_votes = sum(tally.values())
tally["Total Votes: "] = total_votes

print(tally)

```
