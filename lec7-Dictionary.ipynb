
# 04 - Adding and Updating

## Adding 

```python
programming_dictionary = {
    "bugs": "this is an error",
}
programming_dictionary["my name"] = "my name is ethan" #this is how you add a key and value into a existing dictionary

print(programming_dictionary)
```

## Update - Bracket Assignment

```python
programming_dictionary["bugs"] = "the bugs fixed value"#this is how you modify a existing key

print(programming_dictionary)
```

## Update - with update() method

```python
d = {
    "a": 1,
    "b": 2
    }
d.update({"b": 99, "c": 20})
print(d)

d.update(
        d = 4, 
        e = 5
        )
print(d)
```

## Update - setdefualt() method

```python
d = {
    "a":1,
    "b":2
    }

val = d.setdefault("a", 999) #to check if that key exist and if not, it will create a key:value
print(val)
print(d)



val = d.setdefault("c", 3)
print(val)
print(d)
```

# 05 - Removing items

# 06 - iterate over dictionaries

looping over keps

```python
programming_dictionary = {
    "bug": "this is a error",
    "function": "A piece of reusable code",
    "Loop": "Keep looping and doing something",
}

for k in programming_dictionary:
    print(k)
```

```python
programming_dictionary = {
    "bug": "this is a error",
    "function": "A piece of reusable code",
    "Loop": "Keep looping and doing something",
}

for k in programming_dictionary:
    print(programming_dictionary[k]) #instead of printing the key, it prints the value
```

```python
programming_dictionary = {
    "bug": "this is a error",
    "function": "A piece of reusable code",
    "Loop": "Keep looping and doing something",
}

for k in programming_dictionary.values():
    print(k)
```

```python
programming_dictionary = {
    "bug": "this is a error",
    "function": "A piece of reusable code",
    "Loop": "Keep looping and doing something",
}

for k,v in programming_dictionary.items():
    print(f'this is key :{k}, and this is value:{v}')
```

# 07 - How to access a specific key or value

```python
my_dict = {"a": 1, "b": 2, "c": 3}

second_key = list(my_dict.keys())[1] #first it takes in all the keys, ["a", "b","c"] into a list, and then taking the second key which would be [1] and then storing it into the varieble
second_value = list(my_dict.values())[1] #same thing but you only get all the values

print(second_key)    # b
print(second_value)  # 2
```

## copy() -- Shallow Copy

```python
original = {"a": 1, "b": [2,3]}

shallow = original.copy()

shallow["a"] = 999
print(original["a"])

shallow["b"].append(4)
print(original["b"])
```
