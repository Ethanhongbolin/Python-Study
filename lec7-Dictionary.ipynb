# 09. Iterating Over Dictionaries
## Looping Over Keys

```python
programming_dictionary = {
    "Bug": "An error in a program",
    "Function": "A reusable piece of code",
    "Loop": "Doing something repeatedly"
                         }
for key in programming_dictionary:
    print(key)
# Output:
# Bug
# Function
# Loop

```

## Accesing Value

```python
for key in programming_dictionary:
    print(programming_dictionary[key])
```

## Looping Over Values

```python
for value in programming_dictionary.values():
    print(f'these are the values: "{value}"')
# Output:
# An error in a program
# A reusable piece of code
# Doing something repeatedly
```

## Looping Over Key-Value Pairs

```python
for key, value in programming_dictionary.items():
    print(f"{key}: {value}")
# Output:
# Bug: An error in a program
# Function: A reusable piece of code
# Loop: Doing something repeatedly

for name, descritipion in nba_goats.items():
    print(f'{name}: {descritipion}')
```

## Dictionary Unpacking with **

```python
def greet(name, age, city):
    print(f"Hi, I'm {name}, {age} years old, from {city}.")
info = {"name": "Alice","age": 30,"city": "New York"}
greet(**info)


# Output: Hi, I'm Alice, 30 years old, from New York.

```

## Enumerate with Dictionaries

```python
for index, value in enumerate(programming_dictionary.values()):
    print(f"{index}: = {value}")
```

```python
list1 = ['q','w','e']

for index,letters in enumerate(list1):
    print(f'{index}: {letters}')
```

```python
for index, (key, value) in enumerate(programming_dictionary.items()):
    print(f"{index}: {key} = {value}")
# Output:
# 0: Bug = An error in a program
# 1: Function = A reusable piece of code
# 2: Loop = Doing something repeatedly
```

## A4 - Vote Tall

```python
votes = [
    "Alice",
    "Bob",
    "Alice",
    "Charlie",
    "Bob",
    "Alice",
    "Alice",
    "Charlie",
    "Bob",
    "Alice",
    "Charlie",
    "Charlie",
]
tally = {}
for name in votes:
    tally[name] = tally.setdefault(name, 0) + 1
    

total_votes = sum(tally.values())
tally["Total Votes: "] = total_votes

print(tally)

```

## Reproduce A2

```python
stock = {"apples": 12, "bananas": 5, "cherries": 20}
delivery = {"bananas": 10, "dates": 8, "apples": 3}




for fruits in delivery:
    total_stock = delivery[fruits] + stock.get(fruits, 0)
    print(f'{fruits}: {total_stock} ')
    stock[fruits] = stock.get(fruits, 0)
    
print(stock)
```

```python
votes = [
    "Alice",
    "Bob",
    "Alice",
    "Charlie",
    "Bob",
    "Alice",
    "Alice",
    "Charlie",
    "Bob",
    "Alice",
    "Charlie",
    "Charlie",
]

tally = {}
tally["_total"] = 0
for names in votes:
    tally.setdefault(names,0)
    tally[names] = tally[names] + 1
    tally["_total"] += 1
             
    
print(f'{tally}')
```
