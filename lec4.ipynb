# Randomization



```python
import random 
#"random.randint(1,10)" gives a random integer between 1-10, and then stores it into the varieble on the left side, and then it prints it out.
# Random integer between 1-10 (inclusive)
random_inm = random.randint(1,10)
print(random_inm)

# creating random float number between (0-1)
random_float = random.random()
print(random_float)

# Random float between 0-5

random_float = random.random() * 5
print(random_float)



```

# Coin Toss Task:

### Random number, between 0-1, 0 = heads 1 = tails

```python
import random

random_int = random.randint(0,1)
if random_int == 0:
    print("Heads")
else:
    print("Tails")
```

### Q1. Create a list which stores 3 numbers 5, 9, 2 Please find the maximum number and print it out. 
### You are not allowed to use any looping logic such as loop or while loop. No Ai tool is allowed to use to help complete this question.


```python
first_number = float(input("what is your first number"))

second_number = float(input("what is your second number"))

third_number = float(input("what is your third number"))


numbers = [first_number, second_number, third_number]

if numbers[0] - numbers[1] > 0 and numbers[0] - numbers[2]> 0:
        print(f'The number {numbers[0]} is the largest number')

elif numbers[1] - numbers[0] > 0 and numbers[1] - numbers[2]> 0:
        print(f'The number {numbers[1]} is the largest number') 

elif numbers[2] - numbers[1] > 0 and numbers[2] - numbers[1]> 0:
        print(f'The number {numbers[2]} is the largest number')

print(f'These are all the numbers: {numbers}')
```
