# Q4: Grade Checker

# Write a function check_grade(score) that returns:

# "A" if score is 90 or above
# "B" if score is 80 to 89
# "C" if score is 70 to 79
# "D" if score is 60 to 69
# "F" if score is below 60


```python
def check_grade(score):
    if score >= 90:
        print('Your grade is an A')
    if score < 90 and score >= 80:
        print('Your grade is B')
    if score >= 70 and score < 80:
        print('Your grade is C')
    if score >=60 and score < 70:
        print('Your grade is D')
    if score < 60:
        print("F")

grade = check_grade(float(input("what is your grade: ")))
```

# Q5: Count How Many Numbers Are Even

# Write a function count_even(numbers) that receives a list of numbers and returns how many even numbers are inside.

# Example:

# count_even([1, 2, 3, 4, 6]) should return 3


```python
def count_even(number):
    count = 0
    for num in number:
        if num % 2 == 0:
            count = count + 1
    return count

        
set_num = count_even([1,2,3,4,6])

print(set_num)
    
```

# Q6: Find the Largest Number in a List

# Write a function find_largest(numbers) that returns the largest number in a list.

# Example:

# find_largest([5, 8, 2, 9, 3]) should return 9


```python
def find_largest(number):
    largest = number[0]
    for num in number:
        if num > largest:
            largest = num
    
    return largest

set = find_largest([5,8,2,9,3])

print(set)
```

# Q7: Count Passing Students

# Write a function count_passing(scores) that receives a list of scores.
# A student passes if the score is 60 or above.

# The function should return how many students passed.

# Example:

# count_passing([45, 67, 80, 59, 90]) should return 3



```python
def count_passing(score):
    passing = 0
    for num in score:
        if num > 60:
            passing += 1
    return passing

grade = count_passing([45, 67, 80, 59, 90])

print(grade)
```
