# Q2: Write a function find_bigger(a, b) that returns the bigger number.

# Example:

# find_bigger(7, 4) should return 7



```python
def find_bigger(a,b):
    if a > b:
        bigger = a
        return bigger
    if b > a:
        bigger = b
        return bigger

two_num = find_bigger(3,22)
print(two_num)
```
