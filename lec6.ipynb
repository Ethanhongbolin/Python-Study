# Q5: Count How Many Numbers Are Even

 Write a function count_even(numbers) that receives a list of numbers and returns how many even numbers are inside.

 Example:

 count_even([1, 2, 3, 4, 6]) should return 3


```python
def count_even(number):
    count = 0
    for num in number:
        if num % 2 == 0:
            count = count + 1
    return count

        
set_num = count_even([1,2,3,4,6])

print(set_num)
    
```

# Q6: Find the Largest Number in a List

 Write a function find_largest(numbers) that returns the largest number in a list.

 Example:

 find_largest([5, 8, 2, 9, 3]) should return 9


```python
def find_largest(number):
    largest = number[0]
    for num in number:
        if num > largest:
            largest = num
    
    return largest

set = find_largest([5,8,2,9,3])

print(set)
```

# Q7: Count Passing Students

 Write a function count_passing(scores) that receives a list of scores.
 A student passes if the score is 60 or above.

 The function should return how many students passed.

 Example:

 count_passing([45, 67, 80, 59, 90]) should return 3



```python
def count_passing(score):
    passing = 0
    for num in score:
        if num > 60:
            passing += 1
    return passing

grade = count_passing([45, 67, 80, 59, 90])

print(grade)
```

# Q8: Create a New List of Bigger Numbers

 Write a function add_five_to_list(numbers) that receives a list of numbers.
 The function should create and return a new list where each number becomes 5 bigger.

 Example:

 add_five_to_list([1, 2, 3]) should return [6, 7, 8]



```python
def add_five_to_list(number):
    
    new_list = []
    for num in number:
        num += 5
        new_list.append(num)
    return new_list

list1 = add_five_to_list([1,2,3])

print(list1)
```

#  *args and **kwargs
## tuple is a special type of list
## Every thing inside the args is a tuple ()
## Every thing inside the kwargs is a dictionary {}

```python
def add_all(*args):
    print(args)
    total = 0
    for num in args:
        total += num
    return total


print(add_all(1, 2, 3))
print(add_all(1, 2, 3, 4, 5))

```

```python
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f'{key}: {value}')
    print(kwargs["name"])
print_info(name = "Alex", age = 25, city= "Victoria")


```

```python
def demo_1(name,*args,**kwargs):
    print(name)
    for i in args:
        print(i)
    print(kwargs) 
    for key, value in kwargs.items(): 
        print(key, value)
demo_1("ethan",
       (2,3,4), (4,5,6), #this goes into the args, and the brackets turns the 3 numbers into a single element, without it, it would be 3 elements
       age = 15, height = "167cm") #this goes into the kwargs
```

🔹 Task

Write a function called:

class_scores
🔹 It must:
Take a class name as a normal argument
Take any number of student scores using *args
Take extra info using **kwargs (like teacher, room, subject)
🔹 What it should do:

When called, it should print:

1. Class name
2. All scores (clean format)
3. Highest score
4. Lowest score
5. Extra info (from kwargs)
🔹 Example call:
class_scores(
    "10A",
    45, 88, 92, 67,
    teacher="Mr Smith",
    subject="Math"
)
🔹 Expected output (format can vary slightly):
Class: 10A
Scores: 45, 88, 92, 67
Highest: 92
Lowest: 45
Extra Info:
teacher: Mr Smith
subject: Math
🔥 Rules
Must use a function
Must use *args
Must use **kwargs
args will automatically become a tuple
You must calculate highest and lowest using Python

```python
def class_scores(class_name,*args, **kwargs):
    print(class_name)
    print(kwargs)
    print(f'This is the class scores. {args}')
    average = sum(args) / len(args)
    print(f'The class average is {average}')
    largest = args[0]
    for num in args:
        if num > largest:
            largest = num
        
    print(f'The highest grade is {largest}')
    smallest = args[0]
    for num in args:
        if num < smallest:
            smallest = num
    print(f'The lowest grade is {smallest}')



class_scores("ethans class",35, 56, 75, 84, 98, 97, 91, 79, 88, teacher = "Mr Lin")
```
