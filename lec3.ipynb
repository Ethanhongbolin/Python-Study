# PROJECT BMI CALCULATOR 2.0

$bmi = Weight(kg) / Height(meter)^2$

input height in meter and weight in kg and print out the calculated BMI

According to the calculated BMI, tell people there health index

[Reference Website](https://www.medilifecenter.com/blog/what-your-bmi-says-about-your-health)

```python




print("Welecome to Ethans BMI calculator")

name = input("What is your name")
weight = float(input(f'what is your weight in kg, {name}'))

height = float(input(f'what is your height in cm, {name}'))
height = height/100
BMI = weight / height**2

BMI = float(BMI)

if BMI <= 0:
    print("ERROR: Please Try Again")
if BMI < 18.5 and BMI > 0:
    print(f'You are underweight, you are {height} meter tall, you weigh {weight} kg, and your bmi value is {round(BMI, 2)}, thanks for using Ethans BMI Calc 2.0')
if BMI >= 18.5 and BMI < 24.9:
    print(f'You are normal, you are {height} meter tall, you weigh {weight} kg, and your bmi value is {round(BMI, 2)}, thanks for using Ethans BMI Calc 2.0')
if BMI >= 24.9 and BMI < 29.9:
    print(f'You are overweight, you are {height} meter tall, you weigh {weight} kg, and your bmi value is {round(BMI, 2)}, thanks for using Ethans BMI Calc 2.0')
if BMI >= 29.9 and BMI < 34.9:
    print(f'You are obese, you are {height} meter tall, you weigh {weight} kg, and your bmi value is {round(BMI, 2)}, thanks for using Ethans BMI Calc 2.0')
if BMI >= 34.9:
    print(f'You are extremely obese, you are {height} meter tall, you weigh {weight} kg, and your bmi value is {round(BMI, 2)}, thanks for using Ethans BMI Calc 2.0')




```

### input 2 number separately, and write a program to determine which number is bigger(they may be equal, and your program should be able to determine this condition as well)

```python
print("Hello, we will check if your first number is either bigger/smaller than your second number")
first_num = float(input("Please type in your first number"))

second_num = float(input("Please type in your second number"))

if first_num - second_num == 0:
    
    print(f'"The first number is equal to the second number". The first number = {first_num}, the second number = {second_num}.')
elif first_num - second_num > 0:
    print(f'"The first number is bigger than the second number". The first number = {first_num}, the second number = {second_num}.')
elif first_num - second_num < 0:
    print(f'"The first number is smaller than the second number". The first number = {first_num}, the second number = {second_num}.')

print("Thanks for playing...")
```

# Q5:
### design a price calculator for your client. Your client will go shopping at shopping mall. Tax is 12% on top of subtotal. If your client purchased over 200 dolalrs, there is 10% discount applied on top of subtotal. If spending over 500 dollars, there is a 20% discount applied to the subtotal. No tipping option. Write a program to determine what is the final cost according to the input subtotal

```python
subtotal = float(input("Hello, what is your subtotal"))

tax = 0.12

if subtotal < 0:
    print("Please try again, and type a valid positive number")
if subtotal > 200 and subtotal <= 500:
    print("A 10 percent discount has applied to your total")
    subtotal_after_discount = subtotal - subtotal * 0.1
    subtotal_after_discount_and_after_tax = subtotal_after_discount + subtotal_after_discount * tax
    print(f'your final total will be ${subtotal_after_discount_and_after_tax}, thank you for shopping')
if subtotal >= 0 and subtotal <= 200:
    print("There is no discount for you today.")
    final_subtotal = subtotal + subtotal * tax
    print(f'your final total will be ${final_subtotal}, thank you for shopping')
if subtotal > 500:
    print("There is a 20 percent discount that will be applied to your total")
    subtotal_after_discount = subtotal - subtotal * 0.2
    subtotal_after_discount_and_after_tax = subtotal_after_discount + subtotal_after_discount * tax
    print(f'your final total will be ${subtotal_after_discount_and_after_tax}, thank you for shopping')
```
