# PROJECT BMI CALCULATOR 2.0

$bmi = Weight(kg) / Height(meter)^2$

input height in meter and weight in kg and print out the calculated BMI

According to the calculated BMI, tell people there health index

[Reference Website](https://www.medilifecenter.com/blog/what-your-bmi-says-about-your-health)

```python
print("welecome to Ethans BMI calculator")
weight = float(input("what is your weight in kg"))
height = float(input("what is your height in cm"))
height = height/100
BMI = weight / height**2
if BMI < 0:
    print("ERROR: Please Try Again")
if BMI < 18.5 and BMI > 0:
    print("You are underweight")
if BMI > 18.5 and BMI < 24.9:
    print("You are normal")
if BMI > 25 and BMI < 29.9:
    print("You are overweight")
if BMI > 30 and BMI < 34.9:
    print("You are obese")
if BMI > 35:
    print("You are extremely obese")

BMI = float(BMI)

print("your BMI value is", round(BMI, 2))
```
