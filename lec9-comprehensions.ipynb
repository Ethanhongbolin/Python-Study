# Why Comprehension

```python
square = []
for x in range(10):
    square.append(x**2)
print(square)

square = [x ** 2 for x in range(10)]
print(square)
```

# 01 - List Comprehensions

## Basic Syntax
[expression for item in iterable]

```python
number = [1,2,3,4,5]
doubled = []
for n in number:
    doubled.append(n * 2)
print(doubled)

doubled = [n * 2 for n in number]
print(doubled) 

```

## With a Condition (Filtering)
[expression for item in iterable if condition]

```python
evens = []
for x in range(10):
    if x % 2 == 0:
        evens.append(x)
print(evens)
evens = [x for x in range(10) if x % 2 == 0]
print(evens)
```

## What if=else (Condional Expression)
When you need if-else, the conditional goes before the `for` keyword (not after it)

```python
labels = []
for x in range(5):
    if x % 2 == 0:
        labels.append("even")
    else:
        labels.append("odd")
print(labels)
labels = ["even" if x % 2 == 0 else "odd" for x in range(5)]
print(labels)
```

```python
even_odd = [
    "even" if x % 2 == 0
    else "odd" 
    for x in range(5) 
]
print(even_odd)



```

# FizzBuzz Challenge

```python
FizzBuzz = [
    "FizzBuzz" if x % 2 == 0 and x % 3 == 0
    else "even" if x % 2 == 0
    else "odd" 
    for x in range (1,100)
]
print(FizzBuzz)
```

# A1 -  Menu Cleanup

```python
dishes = [
    "  cheese burger",
    "SODA",
    "french Fries  ",
    "  ice cream",
    "GRILLED chicken"
]


menu = [
    food.strip().title() for food in dishes
]
print(menu)

```

# A2 - Sensor Alerts

```python
temperatures = [72, 85, 91, 68, 77, 95, 60, 88]


hot = [hot_days > 80 for hot_days in temperatures]
labels = [
    "HIGH" if weather >= 90
    else "WARM" if weather < 90 and weather > 75
    else "COOL"
    for weather in temperatures
]
print(hot)
print(labels)
```

# A3 - Email Domain Extraction

```python
emails = [
    "alice@gmail.com",
    "bob@yahoo.com",
    "carol@company.org",
    "dave@gmail.com",
    "erin@school.edu"
]
domains = [email for email in emails if email.split("@")[-1] != "gmail.com"]
com_email = [email for email in emails if email.endswith(".com")]
print(domains)
print(com_email)
```
