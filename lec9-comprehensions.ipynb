# Why Comprehension

```python
square = []
for x in range(10):
    square.append(x**2)
print(square)

square = [x ** 2 for x in range(10)]
print(square)
```

# 01 - List Comprehensions

## Basic Syntax
[expression for item in iterable]

```python
number = [1,2,3,4,5]
doubled = []
for n in number:
    doubled.append(n * 2)
print(doubled)

doubled = [n * 2 for n in number]
print(doubled) 

```

## With a Condition (Filtering)
[expression for item in iterable if condition]

```python
evens = []
for x in range(10):
    if x % 2 == 0:
        evens.append(x)
print(evens)
evens = [x for x in range(10) if x % 2 == 0]
print(evens)
```

## What if=else (Condional Expression)
When you need if-else, the conditional goes before the `for` keyword (not after it)

```python
labels = []
for x in range(5):
    if x % 2 == 0:
        labels.append("even")
    else:
        labels.append("odd")
print(labels)
labels = ["even" if x % 2 == 0 else "odd" for x in range(5)]
print(labels)
```

```python
even_odd = [
    "even" if x % 2 == 0
    else "odd" 
    for x in range(5) 
]
print(even_odd)



```

# 02 - Dictionary Comprehension

## Basic Syntax
{key_expr: val_expr for item in iterable}

```python
# loop version
squares = {}
for x in range(5):
    squares[x] = x ** 2
print(squares)

square = {x: x**2 for x in range(5)}
print(squares)
```

## Character Position

```python
s = "code"
character_position = {i:c for c, i in enumerate(s)}
print(character_position)
```

## Filtering With Condition (Dictionary)

```python
even_square = {x: x**2 for x in range(0,10) if x % 2 == 0}
print(even_square)
```

# A3 - Email Domain Extraction

```python
emails = [
    "alice@gmail.com",
    "bob@yahoo.com",
    "carol@company.org",
    "dave@gmail.com",
    "erin@school.edu"
]
domains = [email.split("@")[-1] for email in emails]
com_email = [email for email in emails if email.endswith(".com")]
print(domains)
print(com_email)
```

# A4 - Alternating Caps

```python
sentence = "the quick brown fox jumps over the lazy dog"

for index, word in enumerate(sentence.split(" ")):
    if index % 2 == 0:
        print(word.upper())
    else:
        print(word)

    



# print(words)
```

```python
sentence = "the quick brown fox jumps over the lazy dog"

new_sentence = [word.upper() if index % 2 == 0 else word for index, word in enumerate(sentence.split(" "))]
print(new_sentence)
```
