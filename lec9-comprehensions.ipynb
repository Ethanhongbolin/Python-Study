# Why Comprehension

```python
square = []
for x in range(10):
    square.append(x**2)
print(square)

square = [x ** 2 for x in range(10)]
print(square)
```

# 01 - List Comprehensions

## Basic Syntax
[expression for item in iterable]

```python
number = [1,2,3,4,5]
doubled = []
for n in number:
    doubled.append(n * 2)
print(doubled)

doubled = [n * 2 for n in number]
print(doubled) 

```

## With a Condition (Filtering)
[expression for item in iterable if condition]

```python
evens = []
for x in range(10):
    if x % 2 == 0:
        evens.append(x)
print(evens)
evens = [x for x in range(10) if x % 2 == 0]
print(evens)
```

## What if=else (Condional Expression)
When you need if-else, the conditional goes before the `for` keyword (not after it)

```python
labels = []
for x in range(5):
    if x % 2 == 0:
        labels.append("even")
    else:
        labels.append("odd")
print(labels)
labels = ["even" if x % 2 == 0 else "odd" for x in range(5)]
print(labels)
```

```python
even_odd = [
    "even" if x % 2 == 0
    else "odd" 
    for x in range(5) 
]
print(even_odd)



```

# 02 - Dictionary Comprehension

## Basic Syntax
{key_expr: val_expr for item in iterable}

```python
# loop version
squares = {}
for x in range(5):
    squares[x] = x ** 2
print(squares)

square = {x: x**2 for x in range(5)}
print(squares)
```

## Character Position

```python
s = "code"
character_position = {i:c for c, i in enumerate(s)}
print(character_position)
```

## Filtering With Condition (Dictionary)

```python
even_square = {x: x**2 for x in range(0,10) if x % 2 == 0}
print(even_square)
```

## Transforming Existing Dictionaries

```python
prices = {"apple": 1.20, "banana": 0.50, "cherry": 2.00}

discounted = {}
for fruit, price in prices.items():
    discounted[fruit] = round(price * 0.9, 2)

discounted = {fruit: round(price * 0.9, 2) for fruit, price in prices.items()}

print(discounted)
```

## In Class Exercise

## Swapping Keys and Values


```python
original = {"a": 1,"b": 2,"c": 3}
swapped = {v: k for k, v in original.items()}
print(swapped)
```

## Building a Dictionary from Parallel Lists

```python
names = ["Alice", "Bob", "Charlie"]
ages = ["30", "25", "40"]
people = {name: age for name, age in zip(names, ages)}
print(people)
```

## Conditional Value Assignment

```python
scores = {"Alice": 85, "Bob": 42, "Charlie": 90, "Diana": 55}
results = {name: ("Pass" if score >= 60 else "Fail") for name, score in scores.items()}
print(results)
```

# A4 - Alternating Caps

```python
sentence = "the quick brown fox jumps over the lazy dog"

for index, word in enumerate(sentence.split(" ")):
    if index % 2 == 0:
        print(word.upper())
    else:
        print(word)

    



# print(words)
```

# A6 - Student Grade Anaslyzer

```python
students = {
    "Alice": [88, 92, 79],
    "Bob": [65,70, 58],
    "Charlie": [95, 100, 98],
    "David": [72, 69, 74],
    "Eva": [90, 91, 93]
}
students_average = {name: (sum(scores))/3 for name, scores in students.items()}
print(f'These are the average scores: {students_average}')

higher_than_90 = [(name,average) for name, average in students_average.items() if average > 90]
print(higher_than_90)

grade_convert_to_letter = [f"{name}: A" if average >= 90 else f"{name}: B" if average < 90 and average >= 75 else f"{name}: C" if average < 75 and average >= 60 else f"{name}: F" for name, average in students_average.items()]
print(grade_convert_to_letter)
```
