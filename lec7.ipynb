# Creating Dictionaries
key : value

```python
programming_dictionary = {
    # key : value
    "bug": "this is a error",
    "function": "A piece of reusable code",
    "Loop": "Keep looping and doing something",
}


print(programming_dictionary["function"])#instead of printing using like [0] or [1], a dictionary uses a costomized name as index but not actually a index
```

# 2. Accessing Values


## bracket notation

```python
programming_dictionary = {
    "bug": "this is a error",
    "function": "A piece of reusable code",
    "Loop": "Keep looping and doing something",
}

print(programming_dictionary["function"], programming_dictionary["bug"]) #this is how you add two values together

print(programming_dictionary.keys())
```

## get() method

```python
result = programming_dictionary.get("bug") #result will only take one thing at a timee
print(result)

result = programming_dictionary.get("test")
print(result)

result = programming_dictionary.get("function","bug")
print(result)
```


# 3.Adding and Updating

```python
programming_dictionary = {
    "bugs": "this is an error",
}
programming_dictionary["my name"] = "my name is ethan" #this is how you add a key and value into a existing dictionary

print(programming_dictionary)
```

```python
programming_dictionary["bugs"] = "the bugs fixed value"#this is how you modify a existing key

print(programming_dictionary)
```

# 4.Removing items
del Statement

```python
d = {
    "a": 1, 
    "b" : 2, 
    "c" : 3
}

del d["b"]#delets the specific key and value pair
print(d)#prints the other 2 key and value pair
```

# iterate over dictionaries

looping over keps

```python
programming_dictionary = {
    "bug": "this is a error",
    "function": "A piece of reusable code",
    "Loop": "Keep looping and doing something",
}

for k in programming_dictionary:
    print(k)
```

```python
programming_dictionary = {
    "bug": "this is a error",
    "function": "A piece of reusable code",
    "Loop": "Keep looping and doing something",
}

for k in programming_dictionary:
    print(programming_dictionary[k]) #instead of printing the key, it prints the value
```

```python
programming_dictionary = {
    "bug": "this is a error",
    "function": "A piece of reusable code",
    "Loop": "Keep looping and doing something",
}

for k in programming_dictionary.values():
    print(k)
```

```python
programming_dictionary = {
    "bug": "this is a error",
    "function": "A piece of reusable code",
    "Loop": "Keep looping and doing something",
}

for k,v in programming_dictionary.items():
    print(f'this is key :{k}, and this is value:{v}')
```

# How to access a specific key or value

```python
my_dict = {"a": 1, "b": 2, "c": 3}

second_key = list(my_dict.keys())[1] #first it takes in all the keys, ["a", "b","c"] into a list, and then taking the second key which would be [1] and then storing it into the varieble
second_value = list(my_dict.values())[1] #same thing but you only get all the values

print(second_key)    # b
print(second_value)  # 2
```

# Safe Division (Medium)

```python

def safe_divide(x, y, defualt = 0):
    if y == 0:
        return(defualt)

    result = x / y
    return result

safe_divide(3,0)

```
